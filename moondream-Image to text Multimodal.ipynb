```python
# Create an empty list to store the xlsx files with full paths
import os

csv_files_with_paths = []
path = r"/content/"
# Walk through the directory tree, listing all files and subdirectories
for root, dirs, files in os.walk(path):
    # Iterate through all files in the current directory
    for file in files:
        # Check if the file has the xlsx extension
        if file.endswith('.jpg'):
            # Create the full file path
            full_file_path = os.path.join(root, file)

            # Add the full file path to the list
            csv_files_with_paths.append(full_file_path)

# Print the list of xlsx files with full paths
print(csv_files_with_paths)
print(len(csv_files_with_paths))
```

```python

answers = model.batch_answer(
    images=[Image.open(csv_files_with_paths[0]), Image.open(csv_files_with_paths[1])],
    prompts=["how many people are wearing glass?", "Are they sitting on something?"],
    tokenizer=tokenizer,
)
for i in answers:
  print(i)
```

```python
import numpy as np
import gradio as gr
from PIL import Image
import io

model_id = "vikhyatk/moondream2"
revision = "2024-05-08"
model = AutoModelForCausalLM.from_pretrained(
    model_id, trust_remote_code=True, revision=revision,
    torch_dtype=torch.float16).to("cuda")
tokenizer = AutoTokenizer.from_pretrained(model_id, revision=revision)

def sepia(input_img, question):
    if isinstance(input_img, np.ndarray):
        image = Image.fromarray(input_img)
    else:
        image = Image.open(input_img)


    enc_image = model.encode_image(image)
    answer= model.answer_question(enc_image, question, tokenizer)
    return answer

demo = gr.Interface(sepia, inputs=[
        gr.Image(),
        gr.Textbox(label="Question")
    ], outputs="textbox")

if __name__ == "__main__":
    demo.launch()
```
